```python

import pygol as pygol
import random
import PyILP as pyilp
import itertools
```

```python
const=[]
P, N = pygol.bottom_clause_generation(constant_set = const, positive_example='pos_example.f',
                                negative_example='neg_example.n.f', container = "memory")
```

```python
Train_P, Test_P, Train_N, Test_N=pygol.pygol_train_test_split(test_size=0.4, shuffle=True, positive_file_dictionary=P,
                                                        negative_file_dictionary=N)
```

```python
metrics_aleph = []
metrics_pygol = []
metrics_metagol = []
time_aleph=[]
time_pygol=[]
time_metagol=[]
iteration=20
```

```python
for i in range(1,len(Train_P)+1):
    acc_each_iteration = []
    time_each_iteration = []
    for k in range(0,iteration):
        pos = random.choices(list(Train_P.keys()), k=i)
        pos_dict={}
        #print(pos)
        for j in pos:
            pos_dict[j]=Train_P[j]


        model_aleph= pyilp.aleph_learn( positive_example=list(pos_dict.keys()),
                         negative_example=list(Train_N.keys()), settings="modes.pl", test_size=0)
        time_each_iteration.append(model_aleph.time_learn)
        
        #de= max(len(pos), len(Test_N.keys()))/ ( len(pos) + len(Test_N.keys()) )
        
        if model_aleph.hypothesis:
            metrics_aleph_1 = pygol.evaluate_theory_prolog(model_aleph.hypothesis, "BK.pl", Test_P,Test_N )
            acc_each_iteration.append(metrics_aleph_1.accuracy)
        else:
            acc_each_iteration.append(50)
    metrics_aleph.append(acc_each_iteration)
    time_aleph.append(time_each_iteration)
```

```python
for i in range(1,len(Train_P)+1):
    acc_each_iteration = []
    time_each_iteration = []
    for k in range(0,iteration):
        pos = random.choices(list(Train_P.keys()), k=i)
        pos_dict={}
        #print(pos)
        for j in pos:
            pos_dict[j]=Train_P[j]
        model_pygol = pygol.pygol_learn(pos_dict, Train_N, constant_set= const, max_neg=0, 
                         max_literals=3,   key_size=1,  rule_noise_check=True, recursive=True, verbose=False)
        metrics_pygol_1 = pygol.evaluate_theory_prolog( model_pygol.hypothesis, "BK.pl", Test_P,Test_N )
        acc_each_iteration.append(metrics_pygol_1.accuracy)
        time_each_iteration.append(model_pygol.time_learn)
    metrics_pygol.append(acc_each_iteration)
    time_pygol.append(time_each_iteration)
```

```python

for i in range(1,len(Train_P)+1):
    acc_each_iteration = []
    time_each_iteration = []
    for k in range(0,iteration):
        pos = random.choices(list(Train_P.keys()), k=i)
        pos_dict={}
        #print(pos)
        for j in pos:
            pos_dict[j]=Train_P[j]
        model_metagol=pyilp.metagolnt_learn(file="BK_metagol.pl", positive_example=list(pos_dict.keys()), 
                                             negative_example=list(Train_N.keys()),  shuffle=True,  
                                             test_size=0, num_neg=4, max_run=3)
        de= max(len(pos), len(Test_N.keys()))/ ( len(pos) + len(Test_N.keys()) )
        if model_metagol.hypothesis:
            metrics_metagol_1=pygol.evaluate_theory_prolog(model_metagol.hypothesis,"BK.pl",Test_P, Test_N)
            acc_each_iteration.append(metrics_metagol_1.accuracy)
        else:
            acc_each_iteration.append(de)
        time_each_iteration.append(model_metagol.time_learn)
    metrics_metagol.append(acc_each_iteration)
    time_metagol.append(time_each_iteration)
```

```python
import numpy as np
print("accuracy pygol", np.mean(list(itertools.chain(*metrics_pygol))))
print("Error Accuracy", np.std(list(itertools.chain(*metrics_pygol))))
print("time pygol", np.mean(list(itertools.chain(*time_pygol))))
print("Error Time", np.std(list(itertools.chain(*time_pygol))))
print("----------------------------------------------")
print("accuracy metagol", np.mean(list(itertools.chain(*metrics_metagol))))
print("Error Accuracy", np.std(list(itertools.chain(*metrics_metagol))))
print("time metagol", np.mean(list(itertools.chain(*time_metagol))))
print("Error Time", np.std(list(itertools.chain(*time_metagol))))
print("----------------------------------------------")
print("accuracy aleph", np.mean(list(itertools.chain(*metrics_aleph))))
print("Error Accuracy", np.std(list(itertools.chain(*metrics_aleph))))
print("time aleph", np.mean(list(itertools.chain(*time_aleph))))
print("Error Time", np.std(list(itertools.chain(*time_aleph))))
```

```python
py_acc ={}
al_acc = {}
meta_acc ={}

py_time ={}
al_time = {}
meta_time ={}

py_acc_e ={}
al_acc_e = {}
meta_acc_e ={}

py_time_e ={}
al_time_e = {}
meta_time_e ={}


for i in range(0,len(Train_P)):
    py_acc[i] = np.mean(metrics_pygol[i])
    al_acc[i] = np.mean(metrics_aleph[i])
    meta_acc[i] = np.mean(metrics_metagol[i])
    
    py_acc_e[i] = np.std(metrics_pygol[i])
    al_acc_e[i] = np.std(metrics_aleph[i])
    meta_acc_e[i] = np.std(metrics_metagol[i])
    
    py_time[i] = np.mean(time_pygol[i])
    al_time[i] = np.mean(time_aleph[i])
    meta_time[i] = np.mean(time_metagol[i])
    
    py_time_e[i] = np.std(time_pygol[i])
    al_time_e[i] = np.std(time_aleph[i])
    meta_time_e[i] = np.std(time_metagol[i])
```

```python

print("{:<15} {:<15} {:<15} {:<15} {:<15} ".format("Pos#", "acc" ,"err", "time", "err"))
print("-------------------------------------------------------------------------------")
print("\t\t\t\t PyGol")
print("-------------------------------------------------------------------------------")
for i in range(0,len(Train_P)):
    print("{:<15} {:<15} {:<15} {:<15} {:<15} ".format(i,py_acc[i], py_acc_e[i], round(py_time[i],3), 
                                                       round(py_time_e[i],3) ))
print("-------------------------------------------------------------------------------")
print("\t\t\t\t Metagol")
print("-------------------------------------------------------------------------------")
for i in range(0,len(Train_P)):
    print("{:<15} {:<15} {:<15} {:<15} {:<15} ".format(i,meta_acc[i], meta_acc_e[i], round(meta_time[i],3), 
                                                       round(meta_time_e[i],3) ))
print("-------------------------------------------------------------------------------")
print("\t\t\t\t Aleph")
print("-------------------------------------------------------------------------------")
for i in range(0,len(Train_P)):
    print("{:<15} {:<15} {:<15} {:<15} {:<15} ".format(i,al_acc[i], al_acc_e[i], round(al_time[i],3), 
                                                       round(al_time_e[i],3) ))
```
